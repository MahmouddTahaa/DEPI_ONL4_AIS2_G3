# Abstract Classes

- Abstract classes: can NOT be instantiated
- Methods inside an abstract class may or may not be implemented
- All subclasses of an abstract class must implement all abstractmethods

```python
from abc import ABC, abstractmethod


class Animal(ABC):
    @abstractmethod
    def make_sound(self) -> None:
        pass

    @abstractmethod
    def move(self) -> None:
        pass


class Bird(Animal):
    def make_sound(self) -> None:
        print("Bird noises...")

    def move(self) -> None:
        print("Bird is FLYING")


class Fish(Animal):
    def make_sound(self) -> None:
        print("bo2 bo2 bo2 bo2 bo2.....")

    def move(self) -> None:
        print("Fish is SWIMMING")


class Cat(Animal):
    def make_sound(self) -> None:
        print("Meow....")

    def move(self) -> None:
        print("Cat is WALKING")

# a = Animal() -> error, abstract classes are not instantiatable
b = Bird()
f = Fish()
c = Cat()

print("Bird")
b.make_sound()
b.move()
print("-"*20)
print("Fish")
f.make_sound()
f.move()
print("-"*20)
print("Cat")
c.make_sound()
c.move()
```

- Private Attributes
	- private attrs. are denoted by "__" before their name
	- can't be called outside of the class it's declared in

- Protected Attributes
	- protected attrs. are denoted by a single undersocre before their name
	- can only be used in the class it's declared in and all of its subclasses
	- python doesn't enforce protected attributes, it's just a convention

```python
class MercedesCar:
	__make = "Mercedes" # -> private
	model = "G Class"
	_isElectric = False # -> protected

class MercedesEV(MercedesCar):
	_isElectric = True

c1 = MercedesCar()
# print(c1.__make) # -> error
print(c1.model)

ev = MercedesEV()
print(ev._isElectric) # can be accessed outside its class because protected attrs. are not enforced
```

# Inheritance

```python
class Calculator1:
	def summation(self, n1, n2):
		return n1 + n2

	def subtraction(self, n1, n2):
		return n1 - n2

	def multiplication(self, n1, n2):
		return n1 * n2

	def division(self, n1, n2):
		return n1 / n2


class Calculator2(Calculator1):
	def power(self, n1, n2):
		return n1**n2


c1 = Calculator2()
print(c1.summation(1, 2))
```

```python
# Multiple inheritance (why can't python just learn OOP from java, whyyyyyyyy)
class A:
    def do_this(self):
        print("i am A")


class B(A):
    pass


class C:
    def do_this(self):
        print("i am C")


# ordering of super classes matter (lol).
# do_this() will call do_this() from B -which is inherited from A- not C
class D(B, C):
    pass


d = D()
d.do_this()  # -> i am A
```
